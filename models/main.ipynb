```python
from urllib.parse import urlparse

def hostname_length(url):
    # Ensure that the URL is a string
    if isinstance(url, str):
        return len(urlparse(url).netloc)
    else:
        return 0  # or return None, depending on how you want to handle non-string inputs

# Apply the modified function
df['hostname_length'] = df['url'].apply(lambda i: hostname_length(i))
df
```

```python
def count_www(url):
    # Ensure that the URL is a string
    if isinstance(url, str):
        return url.count('www')
    else:
        return 0  # or return None, depending on how you want to handle non-string inputs

# Apply the modified function
df['count-www'] = df['url'].apply(lambda i: count_www(i))
df
```

```python
def count_https(url):
    # Ensure that the URL is a string
    if isinstance(url, str):
        return url.count('https')
    else:
        return 0  # or return None, depending on how you want to handle non-string inputs

df['count-https'] = df['url'].apply(lambda i: count_https(i))

def count_http(url):
    # Ensure that the URL is a string
    if isinstance(url, str):
        return url.count('http')
    else:
        return 0  # or return None, depending on how you want to handle non-string inputs

df['count-http'] = df['url'].apply(lambda i: count_http(i))

# Display the DataFrame
df
```

```python
import re

def shortening_service(url):
    # Ensure that the URL is a string
    if isinstance(url, str):
        match = re.search(
            'bit\.ly|goo\.gl|shorte\.st|go2l\.ink|x\.co|ow\.ly|t\.co|tinyurl|tr\.im|is\.gd|cli\.gs|'
            'yfrog\.com|migre\.me|ff\.im|tiny\.cc|url4\.eu|twit\.ac|su\.pr|twurl\.nl|snipurl\.com|'
            'short\.to|BudURL\.com|ping\.fm|post\.ly|Just\.as|bkite\.com|snipr\.com|fic\.kr|loopt\.us|'
            'doiop\.com|short\.ie|kl\.am|wp\.me|rubyurl\.com|om\.ly|to\.ly|bit\.do|t\.co|lnkd\.in|'
            'db\.tt|qr\.ae|adf\.ly|goo\.gl|bitly\.com|cur\.lv|tinyurl\.com|ow\.ly|bit\.ly|ity\.im|'
            'q\.gs|is\.gd|po\.st|bc\.vc|twitthis\.com|u\.to|j\.mp|buzurl\.com|cutt\.us|u\.bb|yourls\.org|'
            'x\.co|prettylinkpro\.com|scrnch\.me|filoops\.info|vzturl\.com|qr\.net|1url\.com|tweez\.me|v\.gd|'
            'tr\.im|link\.zip\.net',
            url)
        if match:
            return 1
        else:
            return 0
    else:
        return

# Apply the function to the DataFrame
df['short_url'] = df['url'].apply(lambda i: shortening_service(i))

# Display the DataFrame
df
```

```python
from tld import get_tld

# Adjusting fd_length to handle non-string inputs
from urllib.parse import urlparse

def fd_length(url):
    if isinstance(url, str):  # Check if the URL is a string
        urlpath = urlparse(url).path
        try:
            return len(urlpath.split('/')[1])
        except IndexError:
            return 0
    else:
        return 0  # Handle non-string inputs


def tld_length(tld):
    if isinstance(tld, str):  # Check if the TLD is a string
        return len(tld)
    else:
        return -1  # Handle non-string inputs

from tld import get_tld

df['fd_length'] = df['url'].apply(fd_length)

# Ensure get_tld handles non-strings properly or check before applying
df['tld'] = df['url'].apply(lambda i: get_tld(i, fail_silently=True) if isinstance(i, str) else None)
df['tld_length'] = df['tld'].apply(tld_length)
df
```

```python
import re

def suspicious_words(url):
    # Ensure that the URL is a string
    if isinstance(url, str):
        match = re.search(
            'PayPal|login|signin|bank|account|update|free|lucky|service|bonus|'
            'ebayisapi|webscr',
            url)
        if match:
            return 1
        else:
            return 0
    else:
        return 0  # Return 0 for non-string inputs

# Apply the function to the DataFrame
df['sus_url'] = df['url'].apply(lambda i: suspicious_words(i))



df
```

```python
from urllib.parse import urlparse
import re

def abnormal_url(url):
    # Ensure that the URL is a string
    if isinstance(url, str):
        hostname = urlparse(url).hostname
        hostname = str(hostname)  # Convert hostname to string to ensure it's not None
        match = re.search(hostname, url)
        if match:
            return 1
        else:
            return 0
    else:
        return 0  # Return 0 for non-string inputs

# Apply the function to the DataFrame
df['abnormal_url'] = df['url'].apply(lambda i: abnormal_url(i))
df
```

```python
df['abnormal_url'].value_counts()
```

```python
import re

def having_ip_address(url):
    # Ensure that the URL is a string
    if isinstance(url, str):
        match = re.search(
            '(([01]?\\d\\d?|2[0-4]\\d|25[0-5])\\.([01]?\\d\\d?|2[0-4]\\d|25[0-5])\\.'
            '([01]?\\d\\d?|2[0-4]\\d|25[0-5])\\/)|'  # IPv4
            '((0x[0-9a-fA-F]{1,2})\\.(0x[0-9a-fA-F]{1,2})\\.'
            '(0x[0-9a-fA-F]{1,2})\\.(0x[0-9a-fA-F]{1,2})\\/)'  # IPv4 in hexadecimal
            '(?:[a-fA-F0-9]{1,4}:){7}[a-fA-F0-9]{1,4}',  # Ipv6
            url)
        if match:
            return 1
        else:
            return 0
    else:
        return 0  # Return 0 for non-string inputs

# Apply the function to the DataFrame
df['use_of_ip_address'] = df['url'].apply(lambda i: having_ip_address(i))
df
```

```python
def get_url(url):
    
    status = []
    
    status.append(having_ip_address(url))
    status.append(abnormal_url(url))
    status.append(count_www(url))

    status.append(shortening_service(url))
    status.append(count_https(url))
    status.append(count_http(url))
    
    status.append(url_length(url))
    status.append(hostname_length(url))
    status.append(suspicious_words(url))
    status.append(google_index(url))
    status.append(fd_length(url))
    tld = get_tld(url,fail_silently=True)
      
    status.append(tld_length(tld))

    return status
```

```python
def model_predict5(test_url):
    features_test = get_url(test_url)
    features_test = np.array(features_test).reshape((1, -1))

    pred = classifier5.predict(features_test)
    if int(pred[0]) == 0:
        res="Bening"
        return res
    
    elif int(pred[0]) == 1:
        res="Defacement"
        return res
    
    elif int(pred[0]) == 2:
        res="Phising"
        return res
        
    elif int(pred[0]) == 3:
        res="Malware"
        return res
```
